<h2 style='color:purple' align='center'>Implementation of mini batch grandient descent in python</h2>

Batch gradient descent uses *all* training samples in forward pass to calculate cumulitive error and than we adjust weights using derivaties

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
#         print (m_curr,b_curr, i)
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
```

```python
plt.scatter(x,y,color='red',marker='+')
```

```python
gradient_descent(x,y)
```
